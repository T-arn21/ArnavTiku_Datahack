```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
```

```python
X = merged_df.drop(columns=['respondent_id','xyz_vaccine','seasonal_vaccine'])
y_xyz = merged_df['xyz_vaccine']
y_seasonal = merged_df['seasonal_vaccine']

X_train, X_test, y_xyz_train, y_xyz_test, y_seasonal_train, y_seasonal_test = train_test_split(X, y_xyz, y_seasonal, test_size=0.25, random_state=40)
```

```python
model_xyz = LogisticRegression()
model_xyz.fit(X_train, y_xyz_train)

model_seasonal = LogisticRegression()
model_seasonal.fit(X_train, y_seasonal_train)
```

```python
y_xyz_pred = model_xyz.predict(X_test)
y_seasonal_pred = model_seasonal.predict(X_test)

print("XYZ Vaccine - Accuracy:", accuracy_score(y_xyz_test, y_xyz_pred))
print("Seasonal Vaccine - Accuracy:", accuracy_score(y_seasonal_test, y_seasonal_pred))
```

```python
cv_scores_xyz = cross_val_score(model_xyz, X_train, y_xyz_train, cv=5)
cv_scores_seasonal = cross_val_score(model_seasonal, X_train, y_seasonal_train, cv=5)

print("XYZ Vaccine - CV Scores:", cv_scores_xyz)
print("Seasonal Vaccine - CV Scores:", cv_scores_seasonal)
```

```python
param_grid = {'C': [0.1, 1, 10], 'solver': ['liblinear']}
grid_search = GridSearchCV(LogisticRegression(), param_grid, cv=5)
grid_search.fit(X_train, y_xyz_train)

best_model = grid_search.best_estimator_
```

```python
y_final_xyz_pred = best_model.predict(X_test)
```
