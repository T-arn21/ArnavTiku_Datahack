```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset = pd.read_csv("training_set_features.csv")
df = pd.DataFrame(dataset)
pd.set_option('display.max_columns', None)
df
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df.drop(columns=['employment_occupation','employment_industry','hhs_geo_region'],inplace=True)
```

```python
df.columns
```

```python
df.xyz_concern.fillna(0, inplace=True)
df.xyz_knowledge.fillna(0,inplace=True)
df.behavioral_antiviral_meds.fillna(0,inplace=True)
df.behavioral_avoidance.fillna(0,inplace=True)
df.behavioral_face_mask.fillna(0,inplace=True)
df.behavioral_wash_hands.fillna(0,inplace=True)
df.behavioral_large_gatherings.fillna(0,inplace=True)
df.behavioral_outside_home.fillna(0,inplace=True)
df.behavioral_touch_face.fillna(0,inplace=True)
df.doctor_recc_xyz.fillna(0,inplace=True)
df.doctor_recc_seasonal.fillna(0,inplace=True)
df.chronic_med_condition.fillna(0,inplace=True)
df.child_under_6_months.fillna(0,inplace=True)
df.health_worker.fillna(0,inplace=True)
df.health_insurance .fillna(0,inplace=True)
df.opinion_xyz_vacc_effective.fillna(3,inplace=True)
df.opinion_xyz_risk.fillna(3,inplace=True)
df.opinion_xyz_sick_from_vacc.fillna(3,inplace=True)
df.opinion_seas_vacc_effective.fillna(3,inplace=True)
df.opinion_seas_risk.fillna(3,inplace=True)
df.opinion_seas_sick_from_vacc.fillna(3,inplace=True)
df.isnull().sum()
```

```python
df.education.fillna('< 12 Years',inplace=True)
def education(text):
    if text=='College Graduate':
        return 4
    elif text=='Some College':
        return 3
    elif text=='12 Years':
        return 2
    else:
        return 1
df['education']=df['education'].apply(education)
```

```python
def agegroup(text):
    if '18 - 34' in text:
        return 1
    elif '35 - 44' in text:
        return 2
    elif '45 - 54' in text:
        return 3
    elif '55 - 64' in text:
        return 4
    else:
        return 5

df['age_group']=df['age_group'].apply(agegroup)
```

```python
df=df.loc[df.income_poverty.notnull()]
df=df.loc[df.rent_or_own.notnull() & df.marital_status.notnull() & df.employment_status.notnull()]
df=df.loc[df.household_adults.notnull() & df.household_children.notnull()]
df.isnull().sum()
```

```python
df=pd.get_dummies(df, columns=["race","sex","income_poverty","marital_status","rent_or_own","employment_status","census_msa"])
df
```

```python
df=df.astype('int64')
df.dtypes
```

```python
dataset2 = pd.read_csv("training_set_labels.csv")
df2 = pd.DataFrame(dataset2)
df2
```

```python
df2.isnull().sum()
```

```python
common_column = df.columns[0]
merged_df = pd.merge(df, df2[[common_column, 'xyz_vaccine','seasonal_vaccine']], on=common_column, how='left')
merged_df
```

```python
print(merged_df.seasonal_vaccine.value_counts())
print(merged_df.xyz_vaccine.value_counts())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, roc_auc_score
```

```python
X = merged_df.drop(columns=['respondent_id','xyz_vaccine','seasonal_vaccine'])
y_xyz = merged_df['xyz_vaccine']
y_seasonal = merged_df['seasonal_vaccine']

X_train, X_test, y_xyz_train, y_xyz_test, y_seasonal_train, y_seasonal_test = train_test_split(X, y_xyz, y_seasonal, test_size=0.2, random_state=42)
```

```python
model_xyz1 = LogisticRegression()
model_xyz1.fit(X_train, y_xyz_train)

model_seasonal1 = LogisticRegression()
model_seasonal1.fit(X_train, y_seasonal_train)
```

```python
y_xyz_pred = model_xyz1.predict(X_test)
y_seasonal_pred = model_seasonal1.predict(X_test)

print("XYZ Vaccine - Accuracy:", accuracy_score(y_xyz_test, y_xyz_pred))
print("Seasonal Vaccine - Accuracy:", accuracy_score(y_seasonal_test, y_seasonal_pred))
```

# XYZ VACCINE

```python
model_xyz2 = RandomForestClassifier()
model_xyz2.fit(X_train, y_xyz_train)
y_xyz_pred2 = model_xyz2.predict(X_test)
print("XYZ Vaccine - Accuracy:", accuracy_score(y_xyz_test, y_xyz_pred2))
```

```python
model_xyz3 = KNeighborsClassifier()
model_xyz3.fit(X_train, y_xyz_train)
y_xyz_pred3 = model_xyz3.predict(X_test)
print("XYZ Vaccine - Accuracy:", accuracy_score(y_xyz_test, y_xyz_pred3))
```

```python
model_xyz4 = MLPClassifier()
model_xyz4.fit(X_train, y_xyz_train)
y_xyz_pred4 = model_xyz4.predict(X_test)
print("XYZ Vaccine - Accuracy:", accuracy_score(y_xyz_test, y_xyz_pred4))
```

```python
pred_prob1 = model_xyz1.predict_proba(X_test)
pred_prob2 = model_xyz2.predict_proba(X_test)
pred_prob3 = model_xyz3.predict_proba(X_test)
pred_prob4 = model_xyz4.predict_proba(X_test)
from sklearn.metrics import roc_curve
fpr1, tpr1, thresh1 = roc_curve(y_xyz_test, pred_prob1[:,1], pos_label=1)
fpr2, tpr2, thresh2 = roc_curve(y_xyz_test, pred_prob2[:,1], pos_label=1)
fpr3, tpr3, thresh3 = roc_curve(y_xyz_test, pred_prob3[:,1], pos_label=1)
fpr4, tpr4, thresh4 = roc_curve(y_xyz_test, pred_prob4[:,1], pos_label=1)
random_probs = [0 for i in range(len(y_xyz_test))]
p_fpr, p_tpr, _ = roc_curve(y_xyz_test, random_probs, pos_label=1)


from sklearn.metrics import roc_auc_score

auc_score1 = roc_auc_score(y_xyz_test, pred_prob1[:,1],average='macro')
auc_score2 = roc_auc_score(y_xyz_test, pred_prob2[:,1],average='macro')
auc_score3 = roc_auc_score(y_xyz_test, pred_prob3[:,1],average='macro')
auc_score4 = roc_auc_score(y_xyz_test, pred_prob4[:,1],average='macro')
print(auc_score1,auc_score2,auc_score3,auc_score4)
```

```python
plt.style.use('seaborn')

# plot roc curves
plt.plot(fpr1, tpr1, linestyle='--',color='orange', label='Logistic Regression')
plt.plot(fpr2, tpr2, linestyle='--',color='yellow', label='RFC')
plt.plot(fpr3, tpr3, linestyle='--',color='green', label='KNN')
plt.plot(fpr4, tpr4, linestyle='--',color='red', label='MLPC')
plt.plot(p_fpr, p_tpr, linestyle='--', color='blue')
# title
plt.title('ROC curve')
# x label
plt.xlabel('False Positive Rate')
# y label
plt.ylabel('True Positive rate')

plt.legend(loc='best')
plt.savefig('ROC',dpi=300)
plt.show();
```
